##### 1.1 What languages do developers want to learn in 2024 vs. 2019?

To start, we need to understand developer aspirations. What is the "hot new thing" they are chasing? This is where the stack overflow survey is going to come in clutch

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

# Helper function to process semicolon-delimited strings
def split_and_count(series, delimiter=';'):
    series_cleaned = series.dropna().astype(str)
    all_items = [item.strip() for items_str in series_cleaned for item in items_str.split(delimiter)]
    return Counter(all_items)

# --- Load Data ---
df_2019 = pd.read_csv("major_cleaned_2019_survey.csv")
df_2024 = pd.read_csv("major_cleaned_2024_survey.csv")

# --- Process 2019 Desired Languages ---
lang_2019_counts = split_and_count(df_2019['LanguageDesireNextYear'])
df_lang_2019 = pd.DataFrame(lang_2019_counts.most_common(15), columns=['Language', 'Count_2019'])
df_lang_2019['Percentage_2019'] = (df_lang_2019['Count_2019'] / len(df_2019['LanguageDesireNextYear'].dropna())) * 100

# --- Process 2024 Desired Languages ---
lang_2024_counts = split_and_count(df_2024['LanguageWantToWorkWith'])
df_lang_2024 = pd.DataFrame(lang_2024_counts.most_common(15), columns=['Language', 'Count_2024'])
df_lang_2024['Percentage_2024'] = (df_lang_2024['Count_2024'] / len(df_2024['LanguageWantToWorkWith'].dropna())) * 100

# --- Neutral theme ---
sns.set_theme(style="whitegrid", rc={"axes.edgecolor": "0.8", "grid.color": "0.9"})

# Define subtle neutral palettes
neutral_palette_2019 = sns.color_palette(["#6A8A82", "#A3B18A", "#E9D8A6"])  # muted green-beige tones
neutral_palette_2024 = sns.color_palette(["#588157", "#8DAA91", "#BFD7B5"])  # soft natural greens

plt.figure(figsize=(18, 8))

# --- 2019 Plot ---
plt.subplot(1, 2, 1)
sns.barplot(x='Percentage_2019', y='Language', data=df_lang_2019, palette=neutral_palette_2019)
plt.title('Top 15 Most Desired Languages in 2019', fontsize=16, weight='bold')
plt.xlabel('Percentage of Respondents (%)', fontsize=12)
plt.ylabel('Language', fontsize=12)

# --- 2024 Plot ---
plt.subplot(1, 2, 2)
sns.barplot(x='Percentage_2024', y='Language', data=df_lang_2024, palette=neutral_palette_2024)
plt.title('Top 15 Most Desired Languages in 2024', fontsize=16, weight='bold')
plt.xlabel('Percentage of Respondents (%)', fontsize=12)
plt.ylabel('')

# Clean up layout
sns.despine()
plt.tight_layout()
plt.savefig('desired_languages_2019_vs_2024.png', dpi=300)
print("Charts saved as 'desired_languages_2019_vs_2024.png'")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

# Helper function to process semicolon-delimited strings
def split_and_count(series, delimiter=';'):
    series_cleaned = series.dropna().astype(str)
    all_items = [item.strip() for items_str in series_cleaned for item in items_str.split(delimiter)]
    return Counter(all_items)

# --- Load Data ---
df_2019 = pd.read_csv("major_cleaned_2019_survey.csv")
df_2024 = pd.read_csv("major_cleaned_2024_survey.csv")

# --- Process 2019 Desired Languages ---
lang_2019_counts = split_and_count(df_2019['LanguageDesireNextYear'])
df_lang_2019 = pd.DataFrame(lang_2019_counts.most_common(), columns=['Language', 'Count_2019'])
df_lang_2019['Rank_2019'] = df_lang_2019.index + 1
df_lang_2019['Percentage_2019'] = (df_lang_2019['Count_2019'] / len(df_2019['LanguageDesireNextYear'].dropna())) * 100

# --- Process 2024 Desired Languages ---
lang_2024_counts = split_and_count(df_2024['LanguageWantToWorkWith'])
df_lang_2024 = pd.DataFrame(lang_2024_counts.most_common(), columns=['Language', 'Count_2024'])
df_lang_2024['Rank_2024'] = df_lang_2024.index + 1
df_lang_2024['Percentage_2024'] = (df_lang_2024['Count_2024'] / len(df_2024['LanguageWantToWorkWith'].dropna())) * 100

# --- Merge for rank comparison ---
merged = pd.merge(df_lang_2019, df_lang_2024, on='Language', how='outer')
merged['Rank_Change'] = merged['Rank_2019'] - merged['Rank_2024']
merged['Direction'] = merged['Rank_Change'].apply(lambda x: 'up' if x > 0 else ('down' if x < 0 else 'same'))

# Keep only top 15 languages in 2024
df_2024_top = merged.sort_values('Rank_2024').head(15).copy()

# --- Visualization ---
sns.set_theme(style="whitegrid", rc={"axes.edgecolor": "0.8", "grid.color": "0.9"})

neutral_palette_2019 = sns.color_palette(["#6A8A82", "#A3B18A", "#E9D8A6"])
neutral_palette_2024 = sns.color_palette(["#588157", "#8DAA91", "#BFD7B5"])

plt.figure(figsize=(18, 8))

# --- 2019 Plot ---
plt.subplot(1, 2, 1)
sns.barplot(x='Percentage_2019', y='Language', data=df_lang_2019.head(15), palette=neutral_palette_2019)
plt.title('Top 15 Most Desired Languages in 2019', fontsize=16, weight='bold')
plt.xlabel('Percentage of Respondents (%)', fontsize=12)
plt.ylabel('Language', fontsize=12)

# --- 2024 Plot ---
plt.subplot(1, 2, 2)
sns.barplot(x='Percentage_2024', y='Language', data=df_2024_top, palette=neutral_palette_2024)
plt.title('Top 15 Most Desired Languages in 2024', fontsize=16, weight='bold')
plt.xlabel('Percentage of Respondents (%)', fontsize=12)
plt.ylabel('')

# --- Add up/down arrows beside bars ---
ax = plt.gca()
for i, (lang, direction) in enumerate(zip(df_2024_top['Language'], df_2024_top['Direction'])):
    x_val = df_2024_top['Percentage_2024'].iloc[i]
    y_val = i
    if direction == 'up':
        plt.text(x_val + 0.5, y_val, '', color='green', fontsize=14, va='center', fontweight='bold')
    elif direction == 'down':
        plt.text(x_val + 0.5, y_val, '', color='red', fontsize=14, va='center', fontweight='bold')
    else:
        plt.text(x_val + 0.5, y_val, '', color='gray', fontsize=12, va='center')

sns.despine()
plt.tight_layout()
plt.savefig('figures/desired_languages_2019_vs_2024_arrows.png', dpi=300)
print("Charts saved as 'figures/desired_languages_2019_vs_2024_arrows.png'")
```

##### 1.2: What about the rest of the stack? (Databases & Platforms)
A developer is more than their language. We'll do the same analysis for databases and platforms to see how the "dream stack" has become more cloud-focused.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import re

# --- Helper: Split and count ---
def split_and_count(series, delimiter=';'):
    series_cleaned = series.dropna().astype(str)
    all_items = [item.strip() for items_str in series_cleaned for item in items_str.split(delimiter)]
    return Counter(all_items)

# --- Load Data ---
df_2019 = pd.read_csv("major_cleaned_2019_survey.csv")
df_2024 = pd.read_csv("major_cleaned_2024_survey.csv")

# -------------------------------------------------------------------
# --- DATABASES ---
# -------------------------------------------------------------------
db_2019_counts = split_and_count(df_2019['DatabaseDesireNextYear'])
df_db_2019 = pd.DataFrame(db_2019_counts.most_common(), columns=['Database', 'Count_2019'])
df_db_2019['Rank_2019'] = df_db_2019.index + 1
df_db_2019['Percentage_2019'] = (df_db_2019['Count_2019'] / len(df_2019['DatabaseDesireNextYear'].dropna())) * 100

db_2024_counts = split_and_count(df_2024['DatabaseWantToWorkWith'])
df_db_2024 = pd.DataFrame(db_2024_counts.most_common(), columns=['Database', 'Count_2024'])
df_db_2024['Rank_2024'] = df_db_2024.index + 1
df_db_2024['Percentage_2024'] = (df_db_2024['Count_2024'] / len(df_2024['DatabaseWantToWorkWith'].dropna())) * 100

# Merge to calculate movement
merged_db = pd.merge(df_db_2019, df_db_2024, on='Database', how='outer')
merged_db['Rank_Change'] = merged_db['Rank_2019'] - merged_db['Rank_2024']
merged_db['Direction'] = merged_db['Rank_Change'].apply(lambda x: 'up' if x > 0 else ('down' if x < 0 else 'same'))

# Top 15 for visualization
df_db_2024_top = merged_db.sort_values('Rank_2024').head(15).copy()

# -------------------------------------------------------------------
# --- PLATFORMS ---
# -------------------------------------------------------------------
def clean_aws_2019(x):
    if isinstance(x, str):
        return re.sub(r'\bAWS\b', 'Amazon Web Services (AWS)', x)
    return x

df_2019['PlatformDesireNextYear'] = df_2019['PlatformDesireNextYear'].apply(clean_aws_2019)
plat_2019_counts = split_and_count(df_2019['PlatformDesireNextYear'])
df_plat_2019 = pd.DataFrame(plat_2019_counts.most_common(), columns=['Platform', 'Count_2019'])
df_plat_2019['Rank_2019'] = df_plat_2019.index + 1
df_plat_2019['Percentage_2019'] = (df_plat_2019['Count_2019'] / len(df_2019['PlatformDesireNextYear'].dropna())) * 100

plat_2024_counts = split_and_count(df_2024['PlatformWantToWorkWith'])
df_plat_2024 = pd.DataFrame(plat_2024_counts.most_common(), columns=['Platform', 'Count_2024'])
df_plat_2024['Rank_2024'] = df_plat_2024.index + 1
df_plat_2024['Percentage_2024'] = (df_plat_2024['Count_2024'] / len(df_2024['PlatformWantToWorkWith'].dropna())) * 100

# Merge for rank comparison
merged_plat = pd.merge(df_plat_2019, df_plat_2024, on='Platform', how='outer')
merged_plat['Rank_Change'] = merged_plat['Rank_2019'] - merged_plat['Rank_2024']
merged_plat['Direction'] = merged_plat['Rank_Change'].apply(lambda x: 'up' if x > 0 else ('down' if x < 0 else 'same'))

# Top 15
df_plat_2024_top = merged_plat.sort_values('Rank_2024').head(15).copy()

# -------------------------------------------------------------------
# --- VISUALIZATION STYLE ---
# -------------------------------------------------------------------
sns.set_theme(style="whitegrid", rc={"axes.edgecolor": "0.85", "grid.color": "0.93"})

# Neutral palettes
neutral_db_2019 = sns.color_palette(["#8AA29E", "#C4C5BA", "#E0E0D8"])
neutral_db_2024 = sns.color_palette(["#748CAB", "#A0B2A6", "#D3D3C2"])
neutral_plat_2019 = sns.color_palette(["#A3B18A", "#C8C6A7", "#EDE6D1"])
neutral_plat_2024 = sns.color_palette(["#4C6A77", "#8CA0A0", "#C2C9B0"])

# -------------------------------------------------------------------
# --- DATABASE CHART ---
# -------------------------------------------------------------------
plt.figure(figsize=(18, 8))

# 2019
plt.subplot(1, 2, 1)
sns.barplot(x='Percentage_2019', y='Database', data=df_db_2019.head(15), palette=neutral_db_2019)
plt.title('Top 15 Most Desired Databases in 2019', fontsize=16, weight='bold')
plt.xlabel('Percentage of Respondents (%)', fontsize=12)
plt.ylabel('Database', fontsize=12)

# 2024
plt.subplot(1, 2, 2)
sns.barplot(x='Percentage_2024', y='Database', data=df_db_2024_top, palette=neutral_db_2024)
plt.title('Top 15 Most Desired Databases in 2024', fontsize=16, weight='bold')
plt.xlabel('Percentage of Respondents (%)', fontsize=12)
plt.ylabel('')

# Add arrows
for i, (direction, pct) in enumerate(zip(df_db_2024_top['Direction'], df_db_2024_top['Percentage_2024'])):
    if direction == 'up':
        plt.text(pct + 0.5, i, '', color='green', fontsize=14, va='center', fontweight='bold')
    elif direction == 'down':
        plt.text(pct + 0.5, i, '', color='red', fontsize=14, va='center', fontweight='bold')
    else:
        plt.text(pct + 0.5, i, '', color='gray', fontsize=12, va='center')

sns.despine()
plt.tight_layout()
plt.savefig('figures/desired_databases_2019_vs_2024_arrows.png', dpi=300)
print("Chart saved as 'desired_databases_2019_vs_2024_arrows.png'")

# -------------------------------------------------------------------
# --- PLATFORM CHART ---
# -------------------------------------------------------------------
plt.figure(figsize=(18, 8))

# 2019
plt.subplot(1, 2, 1)
sns.barplot(x='Percentage_2019', y='Platform', data=df_plat_2019.head(15), palette=neutral_plat_2019)
plt.title('Top 15 Most Desired Platforms in 2019', fontsize=16, weight='bold')
plt.xlabel('Percentage of Respondents (%)', fontsize=12)
plt.ylabel('Platform', fontsize=12)

# 2024
plt.subplot(1, 2, 2)
sns.barplot(x='Percentage_2024', y='Platform', data=df_plat_2024_top, palette=neutral_plat_2024)
plt.title('Top 15 Most Desired Platforms in 2024', fontsize=16, weight='bold')
plt.xlabel('Percentage of Respondents (%)', fontsize=12)
plt.ylabel('')

# Add arrows
for i, (direction, pct) in enumerate(zip(df_plat_2024_top['Direction'], df_plat_2024_top['Percentage_2024'])):
    if direction == 'up':
        plt.text(pct + 0.5, i, '', color='green', fontsize=14, va='center', fontweight='bold')
    elif direction == 'down':
        plt.text(pct + 0.5, i, '', color='red', fontsize=14, va='center', fontweight='bold')
    else:
        plt.text(pct + 0.5, i, '', color='gray', fontsize=12, va='center')

sns.despine()
plt.tight_layout()
plt.savefig('figures/desired_platforms_2019_vs_2024_arrows.png', dpi=300)
print("Chart saved as 'desired_platforms_2019_vs_2024_arrows.png'")
```

##### 2.1 What are the most in-demand skills in job postings? (The Plot Twist)
This is where our story gets a major twist. When we analyze the job_skill_set column, we find something surprising.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import ast
from wordcloud import WordCloud

# --- Helper function to process job_skill_set column ---
def parse_and_count_skills(series):
    skills_counter = Counter()
    series_cleaned = series.dropna()
    for item in series_cleaned:
        try:
            skills_list = ast.literal_eval(item)
            if isinstance(skills_list, list):
                cleaned_skills = [skill.strip().lower() for skill in skills_list]
                skills_counter.update(cleaned_skills)
        except (ValueError, SyntaxError):
            continue
    return skills_counter

# --- Load Data ---
df_jobs = pd.read_csv("data/all_job_post.csv")

# --- Filter only IT-related jobs ---
# Adjust column name or filter term if your dataset uses a slightly different format
df_it = df_jobs[df_jobs['category'].str.contains('INFORMATION-TECHNOLOGY', case=False, na=False)]

# --- Parse and count IT job skills ---
job_skills_counts = parse_and_count_skills(df_it['job_skill_set'])
df_job_skills = pd.DataFrame(job_skills_counts.most_common(50), columns=['Skill', 'Count'])

# --- Generate Word Cloud ---
skill_freq = dict(zip(df_job_skills['Skill'], df_job_skills['Count']))

wc = WordCloud(
    width=1400,
    height=900,
    background_color='white',
    colormap='crest',  # soft neutral palette
    max_words=100,
    prefer_horizontal=0.9,
    relative_scaling=0.5,
    min_font_size=10
).generate_from_frequencies(skill_freq)

# --- Plot ---
plt.figure(figsize=(14, 9))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.title('Most In-Demand IT Skills in Job Postings', fontsize=18, weight='bold', pad=20)
plt.tight_layout()
plt.savefig('figures/it_skills_wordcloud.png', dpi=300)
print("Word cloud saved as 'figures/it_skills_wordcloud.png'")
```

##### 3.1: What are the most in-demand technical skills for IT jobs?
This is the core of our analysis. We'll build a "master list" of tech skills from the 2024 survey, then use it to search for demand only within the "INFORMATION-TECHNOLOGY" job category.

```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import seaborn as sns

# --- Step 1: Load your dataset ---
# Example: df = pd.read_csv('your_file.csv')
# Replace this with your actual DataFrame
# It must have a column named 'LanguageHaveWorkedWith'
df = pd.read_csv('major_cleaned_2024_survey.csv')

# --- Step 2: Split and count frequencies ---
def split_and_count(series):
    counter = Counter()
    for row in series.dropna():
        skills = [s.strip().lower() for s in row.split(';') if s.strip()]
        counter.update(skills)
    return counter

skill_counter = split_and_count(df['LanguageHaveWorkedWith'])
df_skill_counts = pd.DataFrame(skill_counter.most_common(), columns=['Skill', 'Count'])

print("Top Skills:")
print(df_skill_counts.head(10))

# --- Step 3A: Visualization - Bar Chart ---
plt.figure(figsize=(10, 6))
sns.barplot(x='Count', y='Skill', data=df_skill_counts.head(20), palette='crest')
plt.title('Top 20 Languages Mentioned in Job Postings', fontsize=15)
plt.xlabel('Number of Postings')
plt.ylabel('Programming Language')
plt.tight_layout()
plt.show()

# --- Step 3B: Visualization - Word Cloud ---
wordcloud = WordCloud(
    width=1000,
    height=600,
    background_color='white',
    colormap='coolwarm',
    max_words=100
).generate_from_frequencies(skill_counter)

plt.figure(figsize=(12, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Programming Language Frequency Word Cloud', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import ast
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

# --- Load dataset ---
df_jobs = pd.read_csv("data/all_job_post.csv")

# --- Filter IT jobs ---
df_it_jobs = df_jobs[df_jobs['category'].str.contains('INFORMATION-TECHNOLOGY', case=False, na=False)]
print(f"Filtered {len(df_it_jobs)} IT job postings out of {len(df_jobs)} total.")

# --- Parse skill lists safely ---
def parse_skills(series):
    counter = Counter()
    for entry in series.dropna():
        try:
            skills = ast.literal_eval(entry)
            if isinstance(skills, list):
                # normalize to lowercase and strip extra spaces
                clean_skills = [s.strip().lower() for s in skills if isinstance(s, str)]
                counter.update(clean_skills)
        except (ValueError, SyntaxError):
            continue
    return counter

# --- Count technical skills ---
skill_counter = parse_skills(df_it_jobs['job_skill_set'])
df_skill_counts = pd.DataFrame(skill_counter.most_common(25), columns=['Skill', 'Count'])

print("Top 10 technical skills in IT job postings:")
print(df_skill_counts.head(10))

# --- Visualization 1: Bar Chart ---
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 8))
sns.barplot(x='Count', y='Skill', data=df_skill_counts, palette='crest')
plt.title('Top 25 Most In-Demand Technical Skills in IT Job Postings', fontsize=16, weight='bold')
plt.xlabel('Number of Job Postings')
plt.ylabel('Technical Skill')
plt.tight_layout()
plt.savefig('figures/top_25_it_technical_skills.png', dpi=300)
plt.show()

# --- Visualization 2: Word Cloud ---
wordcloud = WordCloud(
    width=1400,
    height=900,
    background_color='white',
    colormap='crest',
    max_words=100,
    prefer_horizontal=0.9,
    relative_scaling=0.5,
    min_font_size=10
).generate_from_frequencies(skill_counter)

plt.figure(figsize=(14, 9))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Most In-Demand Technical Skills in IT Job Postings', fontsize=18, weight='bold', pad=20)
plt.tight_layout()
plt.savefig('figures/it_technical_skills_wordcloud.png', dpi=300)
plt.show()

print("Charts saved under /figures/")
```
